```python
import pandas as pd

filePath1 = "purchase_data.json"
filePath2 = "purchase_data2.json"

PurchaseData1 = pd.read_json(filePath1)
PurchaseData2 = pd.read_json(filePath2)

purchaseList =[PurchaseData1,PurchaseData2]

purchaseList = pd.concat(purchaseList)
purchaseList

#Player Count
#---Total Number of Players
totalCount = purchaseList["SN"].nunique()
totalCount = pd.DataFrame({"Total Count":[totalCount]})
totalCount
```

```python
#Purchasing Analysis (Total)
#---Number of Unique Items
uniqueCount = purchaseList["Item ID"].nunique()

#---Average Purchase Price
averagePrice = purchaseList["Price"].mean()

#---Total Number of Purchases
totalPurchases = len(purchaseList)
#---Total Revenue
totalRevenue = purchaseList["Price"].sum()

pd.options.display.float_format = '${:,.2f}'.format
purchasingAnalysis = pd.DataFrame({"Unique Items":[uniqueCount],
                                   "Average Price":[averagePrice],
                                   "Total Purchases":[totalPurchases],
                                  "Total Revenue":[totalRevenue]})

purchasingAnalysis
```

```python
#Gender Demographics
#---Percentage and Count of Male Players
males = purchaseList["Gender"].value_counts()['Male']
malesPercent = (males/totalPurchases)*100

#---Percentage and Count of Female Players
females = purchaseList["Gender"].value_counts()['Female']
femalesPercent = (females/totalPurchases)*100

#---Percentage and Count of Other / Non-Disclosed
other = purchaseList["Gender"].value_counts()['Other / Non-Disclosed']
otherPercent = (other/totalPurchases)*100

pd.options.display.float_format = '{:.2f}%'.format
genderDemo = pd.DataFrame({"Gender":["Males","Females","Other"],
                           "Gender Percentage":[malesPercent,femalesPercent,otherPercent],
                          "Gender Counts":[males,females,other]})

genderDemo
```

```python
#Purchasing Analysis (Gender) --------------------------------------------------NOT DONE!!!!!!!!
#-------------------------------------------------------------------------------NOT DONE!!!!!!!!
#-------------------------------------------------------------------------------NOT DONE!!!!!!!!
#-------------------------------------------------------------------------------NOT DONE!!!!!!!!
#-------------------------------------------------------------------------------NOT DONE!!!!!!!!
#---The below each broken by gender
#------Purchase Count
pd.options.display.float_format = '${:,.2f}'.format

#------Average Purchase Price
#------Total Purchase Value
#------Normalized Totals
malePrice = purchaseList.loc[purchaseList["Gender"] == "Male"]
maleTotalPrice = malePrice["Price"].sum()
maleNormal = malePrice["Price"].mean()
maleAvePrice = maleTotalPrice/males

femalePrice = purchaseList.loc[purchaseList["Gender"] == "Female"]
femaleTotalPrice = femalePrice["Price"].sum()
femaleNormal = femalePrice["Price"].mean()
femaleAvePrice = femaleTotalPrice/females

otherPrice = purchaseList.loc[purchaseList["Gender"] == "Other / Non-Disclosed"]
otherTotalPrice = otherPrice["Price"].sum()
otherNormal = otherPrice["Price"].mean()
otherAvePrice = otherTotalPrice/other


genderPurchaseAnalysis = pd.DataFrame({"Gender":["Males","Females","Other"],
                                      "Purchase Count":[males,females,other],
                                      "Average Purchase Count":[maleAvePrice,femaleAvePrice,otherAvePrice],
                                      "Total Purchase Amount" :[maleTotalPrice,femaleTotalPrice,otherTotalPrice],
                                      "Normalized Totals":[maleNormal,femaleNormal,otherNormal]})

genderPurchaseAnalysis = genderPurchaseAnalysis[["Gender",
                                                 "Purchase Count",
                                                 "Average Purchase Count",
                                                "Total Purchase Amount",
                                                "Normalized Totals"]]
genderPurchaseAnalysis
```

```python
#Age Demographics
#---The below each broken into bins of 4 years (i.e. <10, 10-14, 15-19, etc.)
ageBrackets = [0,9,14,19,24,29,34,39,200]
ageBracketNames = ["< 10","10 - 14","15 - 19","20 - 24", "25 - 29", "30 - 34","35 - 39", "40 +"]

purchaseList["Age Groups"] = pd.cut(purchaseList["Age"], ageBrackets, labels = ageBracketNames)
ageGroups = purchaseList.groupby("Age Groups")

#------Purchase Count
ageCounts = ageGroups["Age"].count()
ageCounts

#------Average Purchase Price
ageAve = ageGroups["Price"].mean()

#------Total Purchase Value
ageSum = ageGroups["Price"].sum()

#------Normalized Totals
#ageNorm = ageGroups["Price"].norm()
# df_norm = (df - df.mean()) / (df.max() - df.min())
#ageNorm=(ageGroups-ageGroups.mean()/ageGroups.max()-ageGroups.min())

agePurchaseAnalysis = pd.DataFrame({"Age Counts":ageCounts,
                                   "Average Price":ageAve,
                                   "Total Price":ageSum})
agePurchaseAnalysis
```

```python
#Top Spenders
#---Identify the the top 5 spenders in the game by total purchase value, then list (in a table):
#------SN
purchaseListMod = purchaseList
purchaseListMod["Count"] = 1

playersDF = purchaseListMod.groupby("SN")
playersDFS = playersDF.sum()
playersDFS = playersDFS.sort_values("Price", ascending = False)

#------Total Purchase Value
SNTotal = playersDFS["Price"].head()

#------Purchase Count
SNCount = playersDFS["Count"].head()

#------Average Purchase Price
SNAve = SNTotal/SNCount

#---------formatting
SNCount = SNCount.map('{:,.0f}'.format)

playerList = pd.DataFrame({"Total Purchase Price":SNTotal,
                           "Purchase Count":SNCount,
                          "Average Price":SNAve})
playerList
```

```python
#Most Popular Items
#---Identify the 5 most popular items by purchase count, then list (in a table):
#------Item ID
#------Total Purchase Value
pop = purchaseListMod.groupby("Item ID")
pop = pd.DataFrame(pop.sum())
pop = pop.reset_index()

#------Item Name
#------Purchase Count
#------Item Price
itemNames = purchaseListMod.iloc[:,[2,3,4]]

popItem = pd.merge(pop, itemNames, on="Item ID")
popItem = popItem.drop_duplicates(["Item Name"], keep = "last")
popItem = popItem.reset_index()
popItem = popItem.drop("index", 1)
popItem = popItem.drop("Age", 1)

popItem = popItem.sort_values("Count",ascending = False)
popItem = popItem.head()

popItem = popItem.rename(columns={"Count":"Purchase Count",
                                 "Price_x":"Total Purchase Value",
                                 "Price_y":"Item Price"})

popItem = popItem[["Item ID",
                  "Item Name",
                  "Purchase Count",
                  "Item Price",
                  "Total Purchase Value"]]
popItem
```
